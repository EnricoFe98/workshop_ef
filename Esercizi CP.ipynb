```python
#EX3 - Greatest Common Divisor using while
def gcd(x,y):
   while True:
       if x < y:
           z = x
           x = y
           y = z
       R = int(x%y)
       if R == 0:
           return y
       else:
           x = y
           y = R

n1 = input('Give me the first value')
n2 = input('Give me a second value')
print(gcd(int(n1),int(n2)))
```

```python
def recfact(x):
    if x == 1 or x == 0:
        return 1
    else:
        return x * recfact(x-1)

def itfact(x):
    if x == 0 or x == 1:
        return 1
    else:
        result = 1
        for i in range(1,(x)+1):
            result = result * i
        return result

factorial = input('give me a number, I will calculate the factorial')
print(recfact(int(factorial)))
print(itfact(int(factorial)))
```

```python
def createexpLd1(L1,L2):
    LE={}
    M = min(len(L1),len(L2))
    i=0
    while i < M:
        k='item'+str(i)
        LE[k]= L1[i]**L2[i]
        i = i+1

    return LE

print(createexpLd1([1,3,3],[1,2,2]))

#MORE COMPLEX
#key is the base
#if I've just one key the exponentiation is the value
#if I have more elements with the same base in the list
#I will have a list of values in the dictionary
#d[3]=x (3 occurs just one time) otherwhise
#d[2]=[x,y] if 2 occurs 2 times

def createexpLd2(L1,L2):
    LE={}
    M = min(len(L1),len(L2))
    i=0
    while i < M:
        if L1[i] in LE:
            if type([])==type(LE[L1[i]]):
                LE[L1[i]].append(L1[i]**L2[i])
            else:
                LE[L1[i]]=[LE[L1[i]]]
                LE[L1[i]].append(L1[i]**L2[i])
        else:
            LE[L1[i]] = L1[i]**L2[i]
        i = i+1

    return LE

print(createexpLd2([1,3,3],[1,2,2]))
```

```python
#with the class you build an object!
class Point:
    #lo usi all'inizio
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    #lo usi quando fai un print
    def __str__(self):
        return '(' + str(self.x) + ',' + str(self.y)+')'
    #lo usi quando fai uan somma
    def __add__(self, other):
        new=Point()
        new.x = self.x + other.x
        new.y = self.y + other.y
        return new
    #compariamo degli oggetti
    def __eq__(self,other):
        return self.x == other.x and self.y ==other.y
    
    def move(self,a,b,):
        self.x = a
        self.y=b

#self è la variabile che indica in quale oggetto stiamo lavorando.
        
k = Point(9,8)
d = Point(100,100)
s = Point(100,100)
print(id(k))
print(k)
k.move(20,54)
print(k)
print(id(k))
print(d+k)
```

```python
#Exercise 5
import string

def weight(S):
    x = string.ascii_lowercase
    count = 0
    for j in range(len(S)):
        for i in range(len(x)):
            if S[j] == x[i]:
                count += int(i)

    return int(count)

def merge(L):
    flag = False
    while not flag:
        flag = True
        for i in range(len(L)-1):
          #BUBLE SORT
            if weight(L[i]) > weight (L[i+1]):
                L[i],L[1+i] = L[1+i],L[i]
                flag = False
    return L

N = int(input('number of the string'))
List_str=[]
for i in range(N):
    List_str.append(input('Dammi una stringa da inserire in lista'))
print(List_str)
print(merge(List_str))
```
